```python
import numpy as np
from scipy import signal
from matplotlib import image
from scipy import misc
from PIL import Image
from skimage import data, io, filters
import matplotlib.pyplot as plt

import requests 
from io import BytesIO
```

```python
response1 = requests.get('https://i.pinimg.com/564x/11/32/ce/1132cea156494ddfc14c0c56390013d7.jpg')
img1 = Image.open(BytesIO(response1.content)).convert('L')

response2 = requests.get('https://i.pinimg.com/564x/b4/c7/69/b4c76909c26fb93e6402190dab581352.jpg')
img2 = Image.open(BytesIO(response2.content)).convert('L')

img1_arr = np.asarray(img1).copy()
img2_arr = np.asarray(img2).copy()

fig, (img_Xiao_Kazuha, img_Yoimiya) = plt.subplots(1, 2, figsize=(16, 9))

img_Xiao_Kazuha.imshow(img1)
img_Xiao_Kazuha.set_axis_off()

img_Yoimiya.imshow(img2)
img_Yoimiya.set_axis_off()
```

```python
img1_prewitt = filters.prewitt(img1)

img2_prewitt = filters.prewitt(img2)

fig, (img_Xiao_Kazuha, img_Yoimiya) = plt.subplots(1, 2, figsize=(16, 9))

img_Xiao_Kazuha.imshow(img1_prewitt, cmap='gray')
img_Xiao_Kazuha.set_axis_off()

img_Yoimiya.imshow(img2_prewitt, cmap='gray')
img_Yoimiya.set_axis_off()
```

```python
img1_sobel = filters.sobel(img1)

img2_sobel = filters.sobel(img2)

fig, (img_Xiao_Kazuha, img_Yoimiya) = plt.subplots(1, 2, figsize=(16, 9))

img_Xiao_Kazuha.imshow(img1_sobel, cmap='gray')
img_Xiao_Kazuha.set_axis_off()

img_Yoimiya.imshow(img2_sobel, cmap='gray')
img_Yoimiya.set_axis_off()
```

```python
img1_scharr = filters.scharr(img1)

img2_scharr = filters.scharr(img2)

fig, (img_Xiao_Kazuha, img_Yoimiya) = plt.subplots(1, 2, figsize=(16, 9))

img_Xiao_Kazuha.imshow(img1_scharr, cmap='gray')
img_Xiao_Kazuha.set_axis_off()

img_Yoimiya.imshow(img2_scharr, cmap='gray')
img_Yoimiya.set_axis_off()
```

```python
fig, (img_Yoimiya_orig, img_Yoimiya_prewitt, img_Yoimiya_sobel, img_Yoimiya_scharr) = plt.subplots(1, 4, figsize=(24, 11))

img_Yoimiya_orig.imshow(img2, cmap='gray')
img_Yoimiya_orig.set_title('original')
img_Yoimiya_orig.set_axis_off()

img_Yoimiya_prewitt.imshow(img2_prewitt, cmap='gray')
img_Yoimiya_prewitt.set_title('prewitt')
img_Yoimiya_prewitt.set_axis_off()

img_Yoimiya_sobel.imshow(img2_sobel, cmap='gray')
img_Yoimiya_sobel.set_title('sobel')
img_Yoimiya_sobel.set_axis_off()

img_Yoimiya_scharr.imshow(img2_scharr, cmap='gray')
img_Yoimiya_scharr.set_title('scharr')
img_Yoimiya_scharr.set_axis_off()
```

A primeira vista no parece ter muita diferena, mas se olharmos com mais ateno podemos perceber que o prewitt faz com que os traos fiquem mais foscos assim diminuindo os detalhes. O sobel e o scharr so bem parecidos porm o scharr traz as linhas mais fortes porm com menos detalhes, j o sobel ele traz mais detalhes da imagem real.
